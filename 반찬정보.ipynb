```python
import re
def find_nhan(a_all):
    tmp = []
    for i in range(len(a_all)):
        if bool(re.search('[^ㄱ-ㅎ가-힣]', a_all[i])):
            tmp.append([i,a_all[i]])
    print(tmp, len(tmp))
# 한글 제외문자 들어간거 전부 찾아냄
```

```python
def find_all(banchen): # 반찬 수 보여주기, 전체 반찬 리턴
    b_all=[]
    for i in banchen:
        b_all.extend(i)
        b_unq = list(set(b_all))
    print('전체반찬수', len(b_all))
    print('반잔종류수', len(b_unq))
    return b_all
```

```python
def menu_sep(menu): #반찬중에 빼야될거 있는경우
    banchen = []
    for ban in menu:
        if '밥' in ban: #쌀이 들어가 있는경우
            pass
        elif '쌀' in ban:#쌀들아가 있는경우
            pass
        elif '(' in ban:
            pass
        elif ')' in ban:
            pass
        elif '<' in ban:
            pass
        elif ban == '':
            pass
        elif '잔반' in ban: #잔반없는날 있음
            pass
        elif '요구르트' in ban:
            pass
        elif '자기' in ban:
            pass
        elif ':' in ban:
            pass
        elif ('양념장' in ban) or ('현미' in ban) or ('음료' in ban) or ('쌈장' in ban):
            pass
        else:
            banchen.append(ban.replace('-','')) # *과 특수문자 re.sub 활용하여 제어 가능
    return banchen
```

## 점심

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules

te = TransactionEncoder()
te_ary = te.fit(l_banchen).transform(l_banchen)
lunch_te = pd.DataFrame(te_ary, columns=te.columns_)
lunch_te
```

```python
## 지지도 신뢰도
from mlxtend.frequent_patterns import apriori

frequent_itemsets_lunch =apriori(lunch_te,min_support=0.001,use_colnames=True)
print(frequent_itemsets_lunch)
#지지도 10퍼센트란 1000개중
# 3000개중에 지지도가 10퍼라면 100개 의 사과가 있다는 뜻???? 
# 지지도는 얼마나 자주나오나(너무 많이나오면 희소성이 떨어진다)
# 향상도는 얼마나 연관있나
#요구르트 빼라, 김치는 하나로 묶어서 김치
'''
핸드폰 영수증은 4개

지지도는 4/3000

지지도를 0.1로 하면
300개의 영수증에 핸드폰 
항목이 나와야함.

지지도(빈도수임)=>대중적임(많은사람들이 가져감)
향상도(A,B) B값이 꼭 A값에 같이 사감.
연관분석(연관된 자료를 추천, 보려고함)

향상도가 

 '''
```

```python
rules_lunch=association_rules(frequent_itemsets_lunch, metric="lift", min_threshold=2)
rules_lunch.sort_values('lift')
```
