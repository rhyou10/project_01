```python
import pandas as pd

df_raw=pd.read_csv('D:/Workspace/04.Project 01/project_01/data/train.csv')
df_raw.head()
```

```python
df_raw['식사가능인원'] = df_raw['본사정원수'] - df_raw['본사휴가자수'] - df_raw['본사출장자수'] - df_raw['본사시간외근무명령서승인건수'] - df_raw['현본사소속재택근무자수']
df_raw.head()
```

```python
weather = pd.read_csv('D:/Workspace/04.Project 01/project_01/data/진주시_날짜_시간_날씨.csv')
print(len(weather))
weather.head()
```

```python
# 일단 결측치 전체 0으로 처리 후 작업
weather_f = weather.fillna(0)
weather_f.isna().sum()
```

```python
weather_f
```

```python
tmp = []
for x in weather_f['강수량(mm)']:
    if x == 0.0:
        tmp.append(0)
    else:
        tmp.append(1)
weather_f['강수여부'] = tmp
```

```python
weather_f['강수여부'].unique()
```

```python
# 비가 온 데이터 2266개
weather_f[weather_f['강수여부']==1]
```

```python
df_raw['중식결식'] = df_raw['식사가능인원'] - df_raw['중식계']
df_raw['석식결식'] = df_raw['식사가능인원'] - df_raw['석식계']
df_raw.head()
```

```python
# 점심 날씨 데이터
lw = weather_f[['일자', '연', '월', '일', '강수여부']][weather_f['시간']== '12:00']
lw.columns = ['일자', '연', '월', '일', '중식강수여부']
lw.head()
```

```python
# 저녁 날씨 데이터
dw = weather_f[['일자','연', '월', '일','강수여부']][weather_f['시간'] == '18:00']
dw.columns = ['일자', '연', '월', '일', '석식강수여부']
dw.head()
```

```python
df_raw =pd.merge(df_raw, lw, how='left')
df_raw.head()
```

```python
df_raw =pd.merge(df_raw, dw, how='left')
df_raw.head()
```

```python
df_raw.info()
```

```python
df_raw['일자'] = pd.to_datetime(df_raw['일자'])
df_raw
```

```python
cor = pd.read_csv('D:/Workspace/04.Project 01/진주확진자수.csv')
cor.columns = ['일자', '일별확진자수']
cor
```
